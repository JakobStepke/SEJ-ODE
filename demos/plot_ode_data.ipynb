```python
# Load data from ode_data.csv and plot it

import matplotlib.pyplot as plt

data_raw = open("ode_data.csv", "r")

data_x = [] # Data in tuples: (time, x)
data_y = [] # Data in tuples: (time, y)

for line in data_raw:
    line = line.split(",")
    data_x.append((float(line[0]), float(line[1])))
    data_y.append((float(line[0]), float(line[2])))

data_raw.close()

# Plot data
plt.plot([x[0] for x in data_x], [x[1] for x in data_x], label="x")
plt.plot([x[0] for x in data_y], [x[1] for x in data_y], label="y")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.show()
```
